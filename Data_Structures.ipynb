2  Explain the difference between mutable and immutable data types with examples?

13 How do sets handle duplicate values in Python?

14 How does the “in” keyword work differently for lists and dictionaries?
  - The in keyword in Python is used to check for the membership of an element in various types of collections, such as lists, dictionaries, sets, etc. However, the way it behaves differs between lists and dictionaries due to the nature of these data structures.
  1. For Lists:
    - When using in with a list, it checks if the value (not the index or key) is present in the list. The operation will return True if the value exists in the list and False otherwise. The search is performed sequentially, so it has an O(n) time complexity, where n is the number of elements in the list.
  2. For Dictionaries:
    - When using in with a dictionary, the behavior changes based on the context. Specifically, it checks for keys in the dictionary, not values or items.

   Checking for keys: The in keyword checks if a specific key exists in the dictionary.

   Checking for values: If you want to check if a value exists, you must explicitly check the dictionary's .values().

15 Can you modify the elements of a tuple? Explain why or why not.
  - No, you cannot directly modify the elements of a tuple in Python because tuples are immutable, meaning their contents cannot be changed after creation.
  Here's why:
  Immutability: Tuples are designed to be immutable, meaning once created, you cannot add, remove, or change elements within the tuple.
  Purpose: Tuples are useful for representing data that should not be altered, such as coordinates, function return values, or data that needs to be treated as a fixed set.
  Error Handling: Attempting to modify a tuple will result in a TypeError.
  Alternative: If you need a mutable sequence, you should use a list instead.

17  Describe the time complexity of accessing elements in a dictionary.
  - Accessing elements in a dictionary (or hash map) typically has an average time complexity of O(1), meaning it takes constant time regardless of the dictionary's size, due to the use of a hash function for efficient lookups.

18  In what situations are lists preferred over dictionaries?
  - For quick data look-ups, configurations, or caches, favor dictionaries. For ordered collections and sequence operations, such as maintaining a stack or queue, lists are more suitable.

19 Why are dictionaries considered unordered, and how does that affect data retrieval?
  - Dictionaries are considered unordered because they store data as key-value pairs, and the order in which these pairs are inserted is not preserved; they are optimized for fast key-based retrieval, not sequential access. This means that when retrieving data, you cannot rely on the order of keys or values being the same as the order they were added.

20  Explain the difference between a list and a dictionary in terms of data retrieval.
  - In terms of data retrieval, lists access elements by their numerical index (starting from 0), while dictionaries access elements by their unique, associated key.
